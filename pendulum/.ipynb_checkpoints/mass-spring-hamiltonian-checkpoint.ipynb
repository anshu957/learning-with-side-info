```julia
include("plotting_utilities.jl")
include("ode_helper.jl")
using SumOfSquares
using DynamicPolynomials
using MosekTools
using Plots
```

```julia
m = 2.
g = 9.81
ℓ = .1
k = 1
α = k / m

# vector field for the Elastic Pendulum
# q = x, p = m * ẋ
function hamiltonian(p, q)
    return q^2 + α * p^2
end

function vf(ẋ, x)
    return [- α * x, ẋ]
end

x0 = 1.
ẋ0 = -1.
T = 3.
num_samples = 10
ẋ_t, x_t = solve_ode(vf, [ẋ0, x0]; verbose=true, tspan = (0., T), length=num_samples)
```

```julia
function background() 
    Plots.plot()
    Plots.xlims!(-2, 2)
    Plots.ylims!(-2, 2)
end


pyplot(size = (400,400))
background()
plot_vectorfield(vf, (-2., 2., 50), (-2., 2, 50))
PyPlot.scatter(ẋ_t, x_t, color=:red)
```

# Learn a polynomial lagrangian

<img src="polynomial_lagrangian.png">


```julia
@polyvar p q
# p => m * ẋ
# q => x


monomials = [p^i * q^j for i=0:2 for j=0:2 if i+j > 0]

model = SOSModel(with_optimizer(Mosek.Optimizer))
@variable model c[1:size(monomials, 1)]
@variable model γ
H = monomials' * c
                
H(1, 0) - H(p=>1, q=>0)
```

```julia
Hp = differentiate(H, p)
Hq = differentiate(H, q)
eval_U_at_v(U, v) = map(Ui -> U())

p_t = m * ẋ_t
q_t = x_t
q̇_t = ẋ_t
# ṗ = m ẍ = - k x
ṗ_t = - k * x_t


error_term = [
    (q̇_t - Hp.(p_t, q_t))...,
    (ṗ_t - Hq.(p_t, q_t))...
    ]
@constraint model [γ, error_term...] in SecondOrderCone()
0
```

```julia
@objective model Min γ
# normalization constraint

optimize!(model)
value(γ)
```
